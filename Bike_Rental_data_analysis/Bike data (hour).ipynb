```python
pd.get_dummies(df['season'], prefix='season', drop_first=True)
```

```python
df['cnt'] = np.log(df['cnt'])
df_oh = df

def one_hot_encoding(data, column):
    data = pd.concat([data, pd.get_dummies(data[column], prefix=column, drop_first=True)], axis=1)
    data = data.drop([column], axis=1)
    return data

cols = ['season','mnth','hr','holiday','weekday','workingday','weathersit']

for col in cols:
    df_oh = one_hot_encoding(df_oh, col)
```

```python
df_oh
```

```python
X = df_oh.drop(columns=['atemp', 'windspeed', 'casual', 'registered', 'cnt'], axis=1)
```

```python
y = df_oh['cnt']
```

```python
from sklearn.linear_model import LinearRegression, Ridge, HuberRegressor, ElasticNetCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
```

```python
models = [LinearRegression(),Ridge(),HuberRegressor(),ElasticNetCV(),DecisionTreeRegressor(),RandomForestRegressor(),ExtraTreesRegressor(),
         GradientBoostingRegressor()]
```

```python
from sklearn import model_selection
```

```python
def train(model):
    kfold = model_selection.KFold(n_splits=5)
    pred = model_selection.cross_val_score(model, X, y, cv=kfold, scoring='neg_mean_squared_error')
    cv_score = pred.mean()
    print('Model:',model)
    print('CV score:', abs(cv_score))
```

```python
for model in models:
    train(model)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
error = y_test - y_pred
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test, y_pred))
```
